```python
#select slow and fast pokemons
ax = df['Speed'].plot(kind='hist',figsize=(10,5),bins=100)
ax.axvline(df['Speed'].quantile(0.05),color='green')
ax.axvline(df['Speed'].quantile(0.95),color='black')
```

```python
top_5= df['Speed'].quantile(0.95)
bottom_5 = df['Speed'].quantile(0.05)
(bottom_5,top_5)
```

```python
slow_fast_poks = df.loc[(df['Speed']<bottom_5) | (df['Speed']>top_5)]
slow_fast_poks
```
